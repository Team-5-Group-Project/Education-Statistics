# 1) Countries with International students ratio around the world
**The chart below gives Information about countries which are more popular for International students**

```python
import pandas as pd

pd.read_csv("1)host_countries_for_int _students.csv", sep = ";")
host_c = pd.read_csv("1)host_countries_for_int _students.csv", sep = ";")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.figure(figsize = (11,5), dpi = 200)

plt.bar(host_c.countries, host_c.International_students / 10**3, ec = "#2b178f", fc = "#2895a8")
plt.title("International students admission")
plt.xlabel("Countries")
plt.ylabel("No of students(thousands)")
plt.legend(["International students portion"])
plt.xticks(color = "#176d8f")
plt.yticks(color = "#178f83")
plt.grid(axis = "y", color = "#2895a8", linestyle = "-.", linewidth = "0.5")

plt.show()
```

# 2) International students(IS) in United States of America
**As the biggest host to International students, admission numbers for USA are given by countires and years** 

```python
#by countries
import pandas as pd

is_c = pd.read_csv("2)is_usa_by_country.csv", sep = ";")
is_y = pd.read_csv("2)is_usa_by_years.csv", sep = ";")

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize = (9,7), dpi = 200)

plt.barh(is_c.countries, is_c.number_of_students/10**3, label = "International students from:", ec = "#ab9a43", fc = "#43ab64")
plt.grid(axis = "x", color = "#43ab64", linestyle = ":", linewidth = "0.5")
plt.title("Home countries of International students",)
plt.ylabel("Countries")
plt.xlabel("Number of students")
plt.xticks(color = "#176d8f")
plt.yticks(color = "#178f83")
plt.legend()

plt.show()

#by admission in the last decade
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.figure(figsize = (11,5), dpi = 200)

plt.plot(is_y.year, is_y.number_of_students / 10**3, "o-", label = "Admission for International students", color = "maroon") #need value lebels
plt.grid(axis = "both", color = "maroon", linestyle = ":", linewidth = "0.5")
plt.title("Admission of International students in the last decade")
plt.xlabel("Years")
plt.ylabel("Admission numbers")
plt.xticks(color = "#176d8f")
plt.yticks(color = "#178f83")
plt.legend(loc=4)

plt.show()
```

# 3) Top Universities of the world
**The line graph below displays the Information about University ranking by main categories in 2021**

```python
import pandas as pd

uni_rank = pd.read_csv("3)_4)top_10_universities.csv", sep = ";")
```

```python
uni_rank
```

```python
import matplotlib.pyplot as plt #put lot of work on this chart
import pandas as pd
import numpy as np

plt.figure(figsize = (11,7), dpi = 200)

plt.plot(uni_rank.Universities, uni_rank.Overall, "o-", label = "Overall")
plt.plot(uni_rank.Universities, uni_rank.Teaching, "o-", label = "Teaching")
plt.plot(uni_rank.Universities, uni_rank.Research, "o-", label = "Research")
plt.plot(uni_rank.Universities, uni_rank.Citations, "o-", label = "Citations")
plt.plot(uni_rank.Universities, uni_rank.International_outlook, "o-", label = "International_outlook")
plt.plot(uni_rank.Universities, uni_rank.students_per_staff, "o-", label = "Student per staff") #this value will be on the right side as line graph
plt.grid(axis = "both", color = "gray", linestyle = ":", linewidth = "0.5")
plt.title("Top Universities around the world")
plt.xlabel("Universities")
plt.ylabel("Rankings (out of 100)")
plt.xticks(color = "#176d8f")
plt.yticks(color = "#178f83")
plt.legend(loc=4)


plt.show()
```

# 4) Gender and International students ration
**As it is given, this chart below gives details about International students and gender ratio for top 10 Universities**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.figure(figsize = (11,7), dpi = 200)

plt.bar(uni_rank.Universities, uni_rank.Female_ratio, label = "Female") #bar chart
plt.bar(uni_rank.Universities, uni_rank.Male_ratio, label = "Male") #bar chart
plt.bar(uni_rank.Universities, uni_rank.International_students, label = "International students") #this value will be line graph
plt.title("International students & Gender ratio")
plt.xlabel("Universities")
plt.ylabel("Percentage(%)")
plt.xticks(color = "#176d8f")
plt.yticks(color = "#178f83")
plt.legend()

plt.show()
```

# 5) Fluctuation of top three Universities
**In the chart below, rankings for Universities (Oxford, Harvard, Cambridge) with each other in the last decade**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

top_three = pd.read_csv("5)top_three_universities_fluctuation.csv", sep = ";")

plt.figure(figsize = (11,7), dpi = 200)

plt.plot(top_three.years, top_three.University_of_Oxford, "o-", label = "University of Oxford") #put value labels
plt.plot(top_three.years, top_three.Harvard_University, "o-", label = "Harvard University") #put all legends too
plt.plot(top_three.years, top_three.University_of_Cambridge, "o-", label = "University of Cambridge") #combine all three "y axes" into one
plt.grid(axis = "both", color = "gray", linestyle = ":", linewidth = "0.5")
plt.title("Rankings for the last decade")
plt.xlabel("Year")
plt.ylabel("Rankings (out of 100)")
plt.xticks(color = "#176d8f")
plt.yticks(color = "#178f83")
plt.legend()

plt.show()
```
