# 1) Host countries to International students around the world
**The chart below gives Information about countries which are more popular for International students**

```python
import pandas as pd

pd.read_csv("1)host_countries_for_int _students.csv", sep = ";")
host_c = pd.read_csv("1)host_countries_for_int _students.csv", sep = ";")
```

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.figure(figsize = (11,5), dpi = 200)

plt.bar(host_c.countries, host_c.International_students / 10**3, ec = "#2b178f", fc = "#2895a8")
plt.title("International students admission", fontsize=17, weight="bold")
plt.xlabel("Countries")
plt.ylabel("No of students(thousands)")
plt.legend(["International students portion"], shadow=True, fontsize="small")
plt.xticks(color = "#176d8f")
plt.yticks(color = "#178f83")
plt.grid(axis = "y", color = "#2895a8", linestyle = "-.", linewidth = "0.5")

plt.show()
```

# 2) International students(IS) in United States of America
**As the biggest host to International students, admission numbers for USA are given by countires and years** 

```python
#by countries
import pandas as pd

is_c = pd.read_csv("2)is_usa_by_country.csv", sep = ";")
is_y = pd.read_csv("2)is_usa_by_years.csv", sep = ";")

import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize = (9,7), dpi = 200)

plt.barh(is_c.countries, is_c.number_of_students/10**3, label = "International students from:", ec = "#ab9a43", fc = "#43ab64")
plt.grid(axis = "x", color = "#2a22bd", linestyle = ":", linewidth = "0.5")
plt.title("Home countries of International students", fontsize=17, weight="bold")
plt.ylabel("Countries")
plt.xlabel("Number of students")
plt.xticks(color = "#176d8f")
plt.yticks(color = "#178f83")
plt.legend(loc=4, shadow=True, fontsize="small")

plt.show()

#by admission in the last decade
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

plt.figure(figsize = (11,5), dpi = 200)

plt.plot(is_y.year, is_y.number_of_students / 10**3, "o-", label = "Admission for International students", color = "maroon", linewidth = 3)
plt.grid(axis = "both", color = "maroon", linestyle = ":", linewidth = "0.5")
plt.title("Admission of International students in the last decade", fontsize=17, weight="bold")
plt.xlabel("Years")
plt.ylabel("Admission numbers")
plt.xticks(color = "#176d8f")
plt.yticks(color = "#178f83")
plt.legend(loc=4, shadow=True, fontsize="small")

plt.show()
```

# 3) Top Universities of the world
**The line graph below displays the Information about University ranking by main categories in 2021**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

universities = ["Oxford", "Stanford", "Harvard", "Cambridge", "Yale", "Princeton", "Columbia", "Toronto", "Peking", "Melbourne"]
overall = [95.6, 94.9, 94.8, 94, 91.6, 91.5, 86.8, 86, 83.9, 78.9]
teaching = [91.3, 92.2, 94.4, 90.3, 91.9, 88.8, 85.1, 75.4, 89.6, 67.1]
research = [99.6, 96.7, 98.8, 99.2, 93.8, 92.5, 82.9, 90.9, 91.3, 76.3]
citations = [98, 99.9, 99.4, 95.6, 97.9, 98.9, 97.7, 94.5, 75.4, 90.1]
students_per_staff =[11.1, 7.4, 9.3, 11, 6, 8, 5.7, 20, 11.1, 27]

fig = plt.figure(figsize=(11, 7), dpi=200)

ax1 = plt.subplot(1,1,1)
ax1.bar(universities, students_per_staff, label = "Students per staff", color = "#15b1bf")
ax1.set_ylim(0, 50)
plt.legend(loc="center left")

ax2 = ax1.twinx()

ax2.plot(universities, overall, "o-", label = "Overall", linewidth = 3)
ax2.plot(universities, teaching, "o-", label = "Teaching", linewidth = 3)
ax2.plot(universities, research, "o-", label = "Research", linewidth = 3)
ax2.plot(universities, citations, "o-", label = "Citations", linewidth = 3)
ax2.yaxis.grid(color = "#15b1bf", linestyle = ":", linewidth = "0.5")
ax1.set_axisbelow(True)

ax1.spines["top"].set_linewidth(0)
ax2.spines["top"].set_linewidth(0)

plt.title("Top Universities around the world", fontsize=17, weight="bold")
plt.legend()
plt.show()
```

# 5) Fluctuation of top three Universities
