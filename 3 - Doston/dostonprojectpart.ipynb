# 1) Countries with International students ratio around the world
**The chart below gives Information about countries which are more popular for International students**

```python
import pandas as pd

pd.read_csv("1)host_countries_for_int _students.csv", sep = ";")
host_c = pd.read_csv("1)host_countries_for_int _students.csv", sep = ";")
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize = (11,5), dpi = 200)

plt.bar(host_c.countries, host_c.International_students / 10**3)
plt.title("International students admission")
plt.xlabel("Countries")
plt.ylabel("No of students(thousands)")
plt.legend(["International students portion"])

plt.show()
```
