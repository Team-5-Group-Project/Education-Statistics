# Top Universities of the world
**The line graph below displays the Information about University ranking by main categories in 2021**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

universities = ["Oxford", "Stanford", "Harvard", "Cambridge", "Yale", "Princeton", "Columbia", "Toronto", "Peking", "Melbourne"]
overall = [95.6, 94.9, 94.8, 94, 91.6, 91.5, 86.8, 86, 83.9, 78.9]
teaching = [91.3, 92.2, 94.4, 90.3, 91.9, 88.8, 85.1, 75.4, 89.6, 67.1]
research = [99.6, 96.7, 98.8, 99.2, 93.8, 92.5, 82.9, 90.9, 91.3, 76.3]
citations = [98, 99.9, 99.4, 95.6, 97.9, 98.9, 97.7, 94.5, 75.4, 90.1]
students_per_staff =[11.1, 7.4, 9.3, 11, 6, 8, 5.7, 20, 11.1, 27]

fig = plt.figure(figsize=(11, 7), dpi=200)

ax1 = plt.subplot(1,1,1)
ax1.bar(universities, students_per_staff, label = "Students per staff", color = "#15b1bf")
ax1.set_ylim(0, 50)
ax1.set_ylabel("Students", fontsize=11, weight="bold", color="#12134a")
ax1.set_xlabel("Universities", fontsize=11, weight="bold", color="#12134a")
plt.legend(fontsize="small", loc="center left")

#======================================================================================================
for index, student in enumerate(students_per_staff):
    ax1.annotate(student, xy=(index-0.09, student+0.5), fontsize="10", weight="bold", color="#fa3939")
#======================================================================================================

ax2 = ax1.twinx()
ax2.plot(universities, overall, "o-", label = "Overall", linewidth = 3)
ax2.plot(universities, teaching, "o-", label = "Teaching", linewidth = 3)
ax2.plot(universities, research, "o-", label = "Research", linewidth = 3)
ax2.plot(universities, citations, "o-", label = "Citations", linewidth = 3)
ax2.yaxis.grid(color = "#15b1bf", linestyle = ":", linewidth = "0.5")
ax2.set_axisbelow(True)
ax2.set_ylabel("Points (out of 100)", fontsize=11, weight="bold", color="#12134a")

ax1.spines["top"].set_linewidth(0)
ax2.spines["top"].set_linewidth(0)

plt.title("Top Universities around the world", fontsize=17, weight="bold")
plt.legend(fontsize="small")
plt.savefig("Top Universities of the world.jpg")
plt.show()
```

# Gender and International students ratio
**As it is given, this chart below gives details about International students and gender ratio for top 10 Universities**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

univer = ["Oxford", "Stanford", "Harvard", "Cambridge", "Yale", "Princeton", "Columbia", "Toronto", "Peking", "Melbourne"]
y = [41, 23, 25, 38, 20, 23, 39, 22, 16, 48]
y1 = [54, 56, 51, 53, 50, 54, 52, 41, 57, 43]
y2 = [46, 44, 49, 47, 50, 46, 48, 59, 43, 57]

fig = plt.figure(figsize=(11,5), dpi=200)

bar_width = 0.27

ax1 = plt.subplot(1,1,1)
ax1.bar(x, y1, label="Male", width=bar_width)
ax1.bar(x+bar_width, y2, label="Female", width=bar_width)

ax1.plot(x, y, "o-", label = "International students", linewidth = 3, color="#b31414")
ax1.set_ylim(0,75)
ax1.set_axisbelow(True)
ax1.spines["top"].set_linewidth(0)
ax1.spines["right"].set_linewidth(0)

#===================================================================================================
for index, student in enumerate(y1):
    ax1.annotate(student, xy=(index-0.13, student+1), fontsize="10", weight="bold", color="#fa3939")
    
for index, student in enumerate(y2):
    ax1.annotate(student, xy=(index+0.17, student+1), fontsize="10", weight="bold", color="#fa3939")
#===================================================================================================

plt.xticks(x+bar_width/2, univer)
plt.xlabel("Universities", fontsize=11, weight="bold", color="#12134a")
plt.ylabel("Percentage (%)", fontsize=11, weight="bold", color="#12134a")
plt.grid(axis="both", color = "grey", linestyle = ":", linewidth = "0.5")

plt.title("International students & Gender ratio", fontsize=17, weight="bold")
plt.legend(fontsize="small")
plt.savefig("Gender and International students ratio.jpg")
plt.show()
```

# Fluctuation of top three Universities
**In the chart below, rankings for Universities (Oxford, Harvard, Cambridge) with each other in the last decade**

# Host countries to International students around the world
**The chart below gives Information about countries which are more popular for International students per year**

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

host_c = pd.read_csv("International_students_world.csv", sep = ";")

plt.figure(figsize=(11,5), dpi=200)
ax1 = plt.subplot(1,1,1)
ax1.bar(host_c.countries, host_c.International_students/10**3, label = "International students portion", color = "#15b1bf")

ax1.set_ylim(0,1300)
ax1.set_ylabel("International students (1000)", fontsize=11, weight="bold", color="#12134a")
ax1.set_xlabel("Countries", fontsize=11, weight="bold", color="#12134a")
ax1.yaxis.grid(linestyle = "--", linewidth = "0.7")
ax1.set_axisbelow(True)
ax1.spines["top"].set_linewidth(0)
ax1.spines["right"].set_linewidth(0)
#===================================================================================================
for index, yval in enumerate(host_c.International_students):
    ax1.annotate(yval, xy=(index-0.4, yval/10**3+17), fontsize="10", weight="bold", color="#fa3939")
#===================================================================================================
plt.title("International students admission", fontsize=17, weight="bold")
plt.legend(fontsize="small", loc="center right")
plt.savefig("Host countries to International students around the world.jpg")
plt.show()
```

# International students(IS) in United States of America
**As the biggest host to International students, admission numbers for USA are given by countires and years** 

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

df_country = pd.read_csv("IS_USA_country.csv", sep = ";")
df_year = pd.read_csv("IS_USA_years.csv", sep = ";")

plt.figure(figsize = (17,11), dpi = 200)

ax1 = plt.subplot(1,2,1)
ax1.barh(df_country.countries, df_country.number_of_students, label = "International students from:")
ax1.set_ylabel("Home countries of students", fontsize=11, weight="bold", color="#12134a")
ax1.set_xlabel("Students", fontsize=11, weight="bold", color="#12134a")
ax1.spines["top"].set_linewidth(0)
ax1.spines["right"].set_linewidth(0)
ax1.set_title("Home countries of students", fontsize=17, weight="bold")
plt.grid(axis="x", linewidth = "0.7", color="#998817")
ax1.set_axisbelow(True)
plt.legend(loc="center right")

#======================================================================================================
for index, student in enumerate(df_country.number_of_students):
    ax1.annotate(student, xy=(student, index), fontsize="10", weight="bold", color="#fa3939")
#======================================================================================================

ax2 = plt.subplot(1,2,2)
ax2.plot(df_year.year, df_year.number_of_students/10**3, "o-", label = "Number of students", linewidth=3)
ax2.set_ylabel("International students (1000)", fontsize=11, weight="bold", color="#12134a")
ax2.set_xlabel("Year (2011~2020)", fontsize=11, weight="bold", color="#12134a")
ax2.spines["top"].set_linewidth(0)
ax2.spines["right"].set_linewidth(0)
ax2.set_title("Admission in the last decade", fontsize=17, weight="bold")
plt.grid(axis="both", linewidth = "0.7", color="#471515")
plt.legend(loc="center right")

plt.savefig("International Universities in USA.jpg")
plt.show()
```
