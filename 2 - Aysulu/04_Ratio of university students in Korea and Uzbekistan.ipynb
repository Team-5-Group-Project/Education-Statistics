```python
import matplotlib.pyplot as plt

import numpy as np

import csv

from matplotlib import ticker

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
# Importing data

x = []
y1 = []
y2 = []
y3 = []
y4 = []
with open('korea.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=';')
    for row in plots:
        x.append(row[0])
        y1.append(int(row[1]))
        y2.append(int(row[2]))
        y3.append(float(row[3]))
        y4.append(float(row[4]))

# Graph - 1 - Name

plt.figure(figsize=(15, 5))
plt.plot(x,y1, 'o-', label='Number of Students in Korea')
plt.plot(x,y2, 'o-', label='Number of Students in Uzbekistan')

plt.title('Number of Students from 2010 to 2020', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Number of students', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()

# Graph - 2 - Name
        
plt.figure(figsize=(15, 5))
plt.plot(x,y3, 'o-', label='Percentage in Korea')
plt.plot(x,y4, 'o-', label='Percentage in Uzbekistan')

plt.title('Percentage  from 2010 to 2020', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Percentage', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()
```
