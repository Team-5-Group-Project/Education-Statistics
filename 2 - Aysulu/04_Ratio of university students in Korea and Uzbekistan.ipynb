```python
import matplotlib.pyplot as plt

import numpy as np

import pandas as pd

from matplotlib import ticker

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

data_rsku = pd.read_csv("04_korea.csv", sep = ";")
data_rsku.head(999)
```

```python
# Graph - 1 - Name

plt.figure(figsize = (11.5, 5), dpi = 300)
plt.plot(data_rsku.Year, data_rsku.Kor, "o-", label='Korea students')
plt.plot(data_rsku.Year, data_rsku.Uzb, "o-", label='Uzbekistan students')

plt.title('Comparison of number of students in Korea and Uzbekistan', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Number of students', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()

# Graph - 2 - Name
        
plt.figure(figsize = (11.5, 5), dpi = 300)
plt.plot(data_rsku.Year, data_rsku.Uzb_per, "o-", label='Uzbekistan students')
plt.plot(data_rsku.Year, data_rsku.Kor_per, "o-", label='Korea students')

plt.title('Percentage  from 2010 to 2020', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Percentage', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()
```
