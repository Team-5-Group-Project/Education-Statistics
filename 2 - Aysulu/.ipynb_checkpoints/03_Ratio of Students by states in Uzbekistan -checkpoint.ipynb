```python
import matplotlib.pyplot as plt

import numpy as np

import csv

from matplotlib import ticker

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
x = []
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
y6 = []
y7 = []
y8 = []
y9 = []
y10 = []
y11 = []
y12 = []
y13 = []
y14 = []
y15 = []

with open('03_ratio_of_students.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(row[0])
        y1.append(float(row[1]))
        y2.append(float(row[2]))
        y3.append(float(row[3]))
        y4.append(float(row[4]))
        y5.append(float(row[5]))
        y6.append(float(row[6]))
        y7.append(float(row[7]))
        y8.append(float(row[8]))
        y9.append(float(row[9]))
        y10.append(float(row[10]))
        y11.append(float(row[11]))
        y12.append(float(row[12]))
        y13.append(float(row[13]))
        y14.append(float(row[14]))
        y15.append(float(row[15]))

        
# Graph - 1 

        
plt.figure(figsize=(15, 5))
plt.plot(x,y2, 'o-', label='Karakalpakstan Republic')
plt.plot(x,y3, 'o-', label='Andijan')
plt.plot(x,y4, 'o-', label='Bukhara')
plt.plot(x,y5, 'o-', label='Jizzakh')
plt.plot(x,y6, 'o-', label='Kashkadarya')
plt.plot(x,y7, 'o-', label='Navoiy')
plt.plot(x,y8, 'o-', label='Namangan')
plt.plot(x,y9, 'o-', label='Samarkand')
plt.plot(x,y10, 'o-', label='Surkhandarya')
plt.plot(x,y11, 'o-', label='Syrdarya')
plt.plot(x,y12, 'o-', label='Tashkent')
plt.plot(x,y13, 'o-', label='Fergana')
plt.plot(x,y14, 'o-', label='Khorezm')

plt.title('Ratio of students by states of Uzbekistan from 2010 to 2019', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Ratio of students', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()


# Graph - 2

plt.figure(figsize=(15, 5))
plt.plot(x,y15, 'o-', label='Tashkent City')
plt.plot(x,y1, 'o-', label='Uzbekistan')

plt.title('Ratio of students in Uzbekistan and Tashkent city from 2010 to 2019', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Ratio of students', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()
```
