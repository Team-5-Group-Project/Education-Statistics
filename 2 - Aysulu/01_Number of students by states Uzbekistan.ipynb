```python
import matplotlib.pyplot as plt

import numpy as np

import pandas as pd

import csv

from matplotlib import ticker

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
# Importing data

x = []
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
y6 = []
y7 = []
y8 = []
y9 = []
y10 = []
y11 = []
y12 = []
y13 = []
y14 = []
y15 = []

with open('01_stat_num.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(row[0])
        y1.append(int(row[1]))
        y2.append(int(row[2]))
        y3.append(int(row[3]))
        y4.append(int(row[4]))
        y5.append(int(row[5]))
        y6.append(int(row[6]))
        y7.append(int(row[7]))
        y8.append(int(row[8]))
        y9.append(int(row[9]))
        y10.append(int(row[10]))
        y11.append(int(row[11]))
        y12.append(int(row[12]))
        y13.append(int(row[13]))
        y14.append(int(row[14]))
        y15.append(int(row[15]))
        
# Graph - 1 - Name

plt.figure(figsize = (11.5, 5), dpi = 300)
plt.plot(x,y1, 'o-', label='Karakalpakstan Republic')
plt.plot(x,y2, 'o-', label='Andijan')
plt.plot(x,y3, 'o-', label='Bukhara')
plt.plot(x,y4, 'o-', label='Jizzakh')
plt.plot(x,y5, 'o-', label='Kashkadarya')
plt.plot(x,y6, 'o-', label='Navoiy')
plt.plot(x,y7, 'o-', label='Namangan')
plt.plot(x,y8, 'o-', label='Samarkand')
plt.plot(x,y9, 'o-', label='Surkhandarya')
plt.plot(x,y10, 'o-', label='Syrdarya')
plt.plot(x,y11, 'o-', label='Tashkent')
plt.plot(x,y12, 'o-', label='Fergana')
plt.plot(x,y13, 'o-', label='Khorezm')

plt.title('Number of students by states of Uzbekistan from 2010 to 2020', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Number of students', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()

# Graph - 2 - Name
        
plt.figure(figsize = (11.5, 5), dpi = 300)
plt.plot(x,y14, 'o-', label='Tashkent City')
plt.plot(x,y15, 'o-', label='Uzbekistan')

plt.title('Number of students in Uzbekistan and Tashkent city from 2010 to 2020', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Number of students', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()
```
