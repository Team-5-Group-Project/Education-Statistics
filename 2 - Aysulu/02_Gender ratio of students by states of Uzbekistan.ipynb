```python
import matplotlib.pyplot as plt

import numpy as np

import csv

from matplotlib import ticker

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
# Importing data

x = []
y1 = []
y2 = []
y3 = []
y4 = []
y5 = []
y6 = []
y7 = []
y8 = []
y9 = []
y10 = []
y11 = []
y12 = []
y13 = []
y14 = []
y15 = []
y16 = []
y17 = []
y18 = []
y19 = []
y20 = []
y21 = []
y22 = []
y23 = []
y24 = []
y25 = []
y26 = []
y27 = []
y28 = []
y29 = []
y30 = []

with open('gen_stat.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(row[0])
        y1.append(float(row[1]))
        y2.append(float(row[2]))
        y3.append(float(row[3]))
        y4.append(float(row[4]))
        y5.append(float(row[5]))
        y6.append(float(row[6]))
        y7.append(float(row[7]))
        y8.append(float(row[8]))
        y9.append(float(row[9]))
        y10.append(float(row[10]))
        y11.append(float(row[11]))
        y12.append(float(row[12]))
        y13.append(float(row[13]))
        y14.append(float(row[14]))
        y15.append(float(row[15]))
        y16.append(float(row[16]))
        y17.append(float(row[17]))
        y18.append(float(row[18]))
        y19.append(float(row[19]))
        y20.append(float(row[20]))
        y21.append(float(row[21]))
        y22.append(float(row[22]))
        y23.append(float(row[23]))
        y24.append(float(row[24]))
        y25.append(float(row[25]))
        y26.append(float(row[26]))
        y27.append(float(row[27]))
        y28.append(float(row[28]))
        y29.append(float(row[29]))
        y30.append(float(row[30]))
        
# Graph - 1 - Female students progress ratio

plt.figure(figsize=(15, 5))
plt.plot(x,y1, 'o-', label='Karakalpakstan Republic')
plt.plot(x,y2, 'o-', label='Andijan')
plt.plot(x,y3, 'o-', label='Bukhara')
plt.plot(x,y4, 'o-', label='Jizzakh')
plt.plot(x,y5, 'o-', label='Kashkadarya')
plt.plot(x,y6, 'o-', label='Navoiy')
plt.plot(x,y7, 'o-', label='Namangan')
plt.plot(x,y8, 'o-', label='Samarkand')
plt.plot(x,y9, 'o-', label='Surkhandarya')
plt.plot(x,y10, 'o-', label='Syrdarya')
plt.plot(x,y11, 'o-', label='Tashkent')
plt.plot(x,y12, 'o-', label='Fergana')
plt.plot(x,y13, 'o-', label='Khorezm')
plt.plot(x,y14, 'o-', label='Tashkent City')

plt.title('Female students progress ratio by states of Uzbekistan from 2010 to 2020', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Percentage of students', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()

# Graph - 2 - Male students progress ratio

plt.figure(figsize=(15, 5))
plt.plot(x,y16, 'o-', label='Karakalpakstan Republic')
plt.plot(x,y17, 'o-', label='Andijan')
plt.plot(x,y18, 'o-', label='Bukhara')
plt.plot(x,y19, 'o-', label='Jizzakh')
plt.plot(x,y20, 'o-', label='Kashkadarya')
plt.plot(x,y21, 'o-', label='Navoiy')
plt.plot(x,y22, 'o-', label='Namangan')
plt.plot(x,y23, 'o-', label='Samarkand')
plt.plot(x,y24, 'o-', label='Surkhandarya')
plt.plot(x,y25, 'o-', label='Syrdarya')
plt.plot(x,y26, 'o-', label='Tashkent')
plt.plot(x,y27, 'o-', label='Fergana')
plt.plot(x,y28, 'o-', label='Khorezm')
plt.plot(x,y29, 'o-', label='Tashkent City')

plt.title('Male students progress ratio by states of Uzbekistan from 2010 to 2020', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Percentage of students', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()

# Graph - 3 - Female and Male students progress ratio in Uzbekistan
        
plt.figure(figsize=(15, 5))

plt.plot(x,y15, 'o-', label='Female')
plt.plot(x,y30, 'o-', label='Male')

plt.title('Female and Male students progress ratio in Uzbekistan from 2010 to 2020', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Number of students', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()
```
