```python
import matplotlib.pyplot as plt

import numpy as np

import pandas as pd

from matplotlib import ticker

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
data_gs = pd.read_csv("02_gen_stat.csv", sep = ",")  
data_gs.head(999)
```

```python
# Graph - 1 - Female students progress ratio

plt.figure(figsize = (11.5, 5), dpi = 300)
plt.plot(data_gs.Year, data_gs.Kar_f, "o-", label='Karakalpakstan Republic')
plt.plot(data_gs.Year, data_gs.And_f, "o-", label='Andijan')
plt.plot(data_gs.Year, data_gs.Buk_f, "o-", label='Bukhara')
plt.plot(data_gs.Year, data_gs.Jiz_f, "o-", label='Jizzakh')
plt.plot(data_gs.Year, data_gs.Kash_f, "o-", label='Kashkadarya')
plt.plot(data_gs.Year, data_gs.Nav_f, "o-", label='Navoiy')
plt.plot(data_gs.Year, data_gs.Nam_f, "o-", label='Namangan')
plt.plot(data_gs.Year, data_gs.Sam_f, "o-", label='Samarkand')
plt.plot(data_gs.Year, data_gs.Sur_f, "o-", label='Surkhandarya')
plt.plot(data_gs.Year, data_gs.Syr_f, "o-", label='Syrdarya')
plt.plot(data_gs.Year, data_gs.Tash_f, "o-", label='Tashkent')
plt.plot(data_gs.Year, data_gs.Fer_f, "o-", label='Fergana')
plt.plot(data_gs.Year, data_gs.Khor_f, "o-", label='Khorezm')
plt.plot(data_gs.Year, data_gs.Tashs_f, "o-", label='Tashkent City')

plt.title('Female students progress ratio by states of Uzbekistan from 2010 to 2020', fontdict = font1, weight="bold", color = 'midnightblue')
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Percentage of students', fontdict = font2)

plt.legend(loc='upper left')
plt.grid(axis = 'both', color = 'blue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='midnightblue')
plt.show()

# Graph - 2 - Male students progress ratio

plt.figure(figsize = (11.5, 5), dpi = 300)
plt.plot(data_gs.Year, data_gs.Kar_m, "o-", label='Karakalpakstan Republic')
plt.plot(data_gs.Year, data_gs.And_m, "o-", label='Andijan')
plt.plot(data_gs.Year, data_gs.Buk_m, "o-", label='Bukhara')
plt.plot(data_gs.Year, data_gs.Jiz_m, "o-", label='Jizzakh')
plt.plot(data_gs.Year, data_gs.Kash_m, "o-", label='Kashkadarya')
plt.plot(data_gs.Year, data_gs.Nav_m, "o-", label='Navoiy')
plt.plot(data_gs.Year, data_gs.Nam_m, "o-", label='Namangan')
plt.plot(data_gs.Year, data_gs.Sam_m, "o-", label='Samarkand')
plt.plot(data_gs.Year, data_gs.Sur_m, "o-", label='Surkhandarya')
plt.plot(data_gs.Year, data_gs.Syr_m, "o-", label='Syrdarya')
plt.plot(data_gs.Year, data_gs.Tash_m, "o-", label='Tashkent')
plt.plot(data_gs.Year, data_gs.Fer_m, "o-", label='Fergana')
plt.plot(data_gs.Year, data_gs.Khor_m, "o-", label='Khorezm')
plt.plot(data_gs.Year, data_gs.Tashs_m, "o-", label='Tashkent City')

plt.title('Male students progress ratio by states of Uzbekistan from 2010 to 2020', fontdict = font1, weight="bold", color = 'midnightblue')
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Percentage of students', fontdict = font2)

plt.legend(loc="upper left")
plt.grid(axis = 'both', color = 'blue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='midnightblue')
plt.show()

# Graph - 3 - Female and Male students progress ratio in Uzbekistan
        
plt.figure(figsize = (11.5, 5), dpi = 300)
plt.plot(data_gs.Year, data_gs.Uzb_f, "o-", label='Female')
plt.plot(data_gs.Year, data_gs.Uzb_m, "o-", label='Male', color='orange')

plt.title("Female and Male students' progress ratio in Uzbekistan from 2010 to 2020", fontdict = font1, weight="bold", color = 'midnightblue')
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Number of students', fontdict = font2)

plt.legend(loc='center left')
plt.grid(axis = 'both', color = 'blue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='midnightblue')
plt.show()
```
