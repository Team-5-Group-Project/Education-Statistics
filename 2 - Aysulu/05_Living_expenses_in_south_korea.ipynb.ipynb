```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as py

pd.read_csv("05_living_expenses_in_korea.csv", sep = ";")
liv= pd.read_csv("05_living_expenses_in_korea.csv", sep = ";")
```

```python
plt.figure(figsize = (11.5, 5), dpi = 300)

plt.bar(liv.type_of_expenses,liv.cost_in_usd/ 6**2, fc = "palevioletred", ec = "crimson") 
plt.title("Monthly living expenses in South Korea", color="purple", size=15)
plt.xlabel("Type of expenses", color="mediumvioletred", size=12)
plt.ylabel("Cost in USD", color= "mediumvioletred", size=12)
plt.xticks(color = "purple")
plt.yticks(color = "purple")
plt.grid(axis = "y", color = "hotpink", linestyle = "-.", linewidth = "0.5")

plt.show()
```
