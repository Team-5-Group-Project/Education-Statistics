```python
import matplotlib.pyplot as plt

import numpy as np

import pandas as pd

import csv

from matplotlib import ticker

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
data_rs = pd.read_csv("03_ratio_of_students.csv", sep = ",")  
data_rs.head(999)
```

```python
data_rs = pd.read_csv("03_ratio_of_students.csv", sep = ",")  

# Graph - 1 
        
plt.figure(figsize = (11.5, 5), dpi = 300)
plt.plot(data_rs.Year, data_rs.Kar, "o-", label='Karakalpakstan Republic')
plt.plot(data_rs.Year, data_rs.And, "o-", label='Andijan')
plt.plot(data_rs.Year, data_rs.Buk, "o-", label='Bukhara')
plt.plot(data_rs.Year, data_rs.Jiz, "o-", label='Jizzakh')
plt.plot(data_rs.Year, data_rs.Kash, "o-", label='Kashkadarya')
plt.plot(data_rs.Year, data_rs.Nav, "o-", label='Navoiy')
plt.plot(data_rs.Year, data_rs.Nam, "o-", label='Namangan')
plt.plot(data_rs.Year, data_rs.Sam, "o-", label='Samarkand')
plt.plot(data_rs.Year, data_rs.Sur, "o-", label='Surkhandarya')
plt.plot(data_rs.Year, data_rs.Syr, "o-", label='Syrdarya')
plt.plot(data_rs.Year, data_rs.Tash, "o-", label='Tashkent')
plt.plot(data_rs.Year, data_rs.Fer, "o-", label='Fergana')
plt.plot(data_rs.Year, data_rs.Khor, "o-", label='Khorezm')
plt.plot(data_rs.Year, data_rs.Tashs, "o-", label='Tashkent City')

plt.title('Ratio of students by states of Uzbekistan and Tashkent city from 2010 to 2019', fontdict = font1, weight="bold", color = 'midnightblue')
plt.ylabel('Ratio of students', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='midnightblue')
plt.show()


# Graph - 2

plt.figure(figsize = (11.5, 5), dpi = 300)
plt.plot(data_rs.Year, data_rs.Uzb, "o-", label='Uzbekistan', color='royalblue')

plt.title('Ratio of students in Uzbekistan from 2010 to 2019', fontdict = font1, weight="bold", color = 'midnightblue')
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Ratio of students', fontdict = font2)


plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='midnightblue')
plt.show()
```
