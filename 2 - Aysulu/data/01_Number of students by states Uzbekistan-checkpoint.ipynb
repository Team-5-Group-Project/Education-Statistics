```python
import matplotlib.pyplot as plt

import numpy as np

import pandas as pd

from matplotlib import ticker

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
data_sn = pd.read_csv("01_stat_num.csv", sep = ",")  

# Graph - 1 - Name      

plt.figure(figsize = (11.5, 5), dpi = 300)
plt.plot(data_sn.Year, data_sn.Kar, "o-", label='Karakalpakstan Republic')
plt.plot(data_sn.Year, data_sn.And, "o-", label='Andijan')
plt.plot(data_sn.Year, data_sn.Buk, "o-", label='Bukhara')
plt.plot(data_sn.Year, data_sn.Jiz, "o-", label='Jizzakh')
plt.plot(data_sn.Year, data_sn.Kash, "o-", label='Kashkadarya')
plt.plot(data_sn.Year, data_sn.Nav, "o-", label='Navoiy')
plt.plot(data_sn.Year, data_sn.Nam, "o-", label='Namangan')
plt.plot(data_sn.Year, data_sn.Sam, "o-", label='Samarkand')
plt.plot(data_sn.Year, data_sn.Sur, "o-", label='Surkhandarya')
plt.plot(data_sn.Year, data_sn.Syr, "o-", label='Syrdarya')
plt.plot(data_sn.Year, data_sn.Tash, "o-", label='Tashkent')
plt.plot(data_sn.Year, data_sn.Fer, "o-", label='Fergana')
plt.plot(data_sn.Year, data_sn.Khor, "o-", label='Khorezm')

plt.title('Number of students by states of Uzbekistan from 2010 to 2020', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Number of students', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()

# Graph - 2 - Name     

plt.figure(figsize = (11.5, 5), dpi = 300)
plt.plot(data_sn.Year, data_sn.Tashs, "o-", label='Tashkent City')
plt.plot(data_sn.Year, data_sn.Uzb, "o-", label='Uzbekistan')

plt.title('Number of students in Uzbekistan and Tashkent city from 2010 to 2020', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Number of students', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()
```
