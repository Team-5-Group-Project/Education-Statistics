```python
import matplotlib.pyplot as plt

import numpy as np

import csv

from matplotlib import ticker

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
# Importing data

x = []
y1 = []
y2 = []

with open('05_major_salary.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=';')
    for row in plots:
        x.append(row[0])
        y1.append(int(row[1]))
        y2.append(int(row[2]))
        
# Graph - 1 - Name

plt.figure(figsize=(15, 5))
plt.plot(x,y1, 'o-', label='Early Career Median Wage')
plt.plot(x,y2, 'o-', label='Mid-Career Median Wage')

plt.title('Majors by salary', fontdict = font1)
plt.xlabel('Majors', fontdict = font2)
plt.ylabel('Salary in USD', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()
```
