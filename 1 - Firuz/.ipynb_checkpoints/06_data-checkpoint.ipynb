```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

fah= pd.read_csv("06_data.csv", sep = ";")
fah.head(999)


```

```python
plt.figure(figsize = (9,7), dpi = 200)

plt.barh(fah.Country, fah.Percentage, label = "% of 25-34 year-olds with an educational level below upper secondary in selected countries", ec = "dodgerblue", fc = "dodgerblue")
plt.grid(axis = "x", color = "#2a22bd", linestyle = ":", linewidth = "0.5")
plt.title("The Countries Where People Fail To Obtain A High School Qualification ", fontsize=17, weight="bold", color = 'midnightblue')
plt.ylabel("Countries", color= "firebrick", fontsize=12)
plt.xlabel("Number of students", color= "firebrick", fontsize=12)
plt.xticks(color = "midnightblue")
plt.yticks(color = "midnightblue")
plt.legend(loc=4, shadow=True, fontsize="small")

plt.show()
```
