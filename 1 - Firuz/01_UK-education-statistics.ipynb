```python
import matplotlib.pyplot as plt

import pandas as pd

import numpy as np

from matplotlib import ticker

import csv

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
# Importing data

x = []
y1 = []
y2 = []
y3 = []

with open('01_UK-education-statistics.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=',')
    for row in plots:
        x.append(row[0])
        y1.append(int(row[1]))
        y2.append(int(row[2]))
        y3.append(int(row[3]))
        
# Graph - 1 - Name

plt.figure(figsize=(15, 5))
plt.plot(x,y1, 'o-', label='Undergraduate')
plt.plot(x,y2, 'o-', label='Postgraduate')
plt.plot(x,y3, 'o-', label='Total')

plt.title('Number of students in UK from 2000 to 2020', fontdict = font1)
plt.xlabel('Years', fontdict = font2)
plt.ylabel('Number of students', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()
```
