```python
import matplotlib.pyplot as plt

import numpy as np

import csv

from matplotlib import ticker

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}
```

```python
# Importing data

x = []
y1 = []
y2 = []
y3 = []

with open('04_major_underemployment_rate.txt','r') as csvfile:
    plots = csv.reader(csvfile, delimiter=';')
    for row in plots:
        x.append(row[0])
        y1.append(float(row[1]))
        y2.append(float(row[2]))
        y3.append(float(row[3]))
        
# Graph - 1 - Name

plt.figure(figsize=(15, 5))
plt.plot(x,y1, 'o-', label='Unemployment rate')
plt.plot(x,y2, 'o-', label='Underemployment rate')
plt.plot(x,y3, 'o-', label='Share with Graduate Degree')

plt.title('Graduate unemployment', fontdict = font1)
plt.xlabel('Majors', fontdict = font2)
plt.ylabel('Percentage', fontdict = font2)

plt.legend()
plt.grid(axis = 'both', color = 'lightblue', linestyle = ':', linewidth = 0.5)
plt.tick_params(axis ='both', which ='major', labelsize = 10, pad = 10, colors ='g')
plt.show()
```
